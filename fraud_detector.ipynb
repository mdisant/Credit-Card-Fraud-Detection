Data Pre-Processing

```python
# Separating the Data
legit = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
print(f"Legit shape: {legit.shape}")
print(f"Fraud shape: {fraud.shape}")
```

```python
# Statistical Measures of the Data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
cc_data.groupby('Class').mean()
```

```python
# Build a sample dataset containing similar distribution of legit and fraud transactions
legit_sample = legit.sample(n=492)
```

```python
# Concatentating two DataFrames
new_cc_data = pd.concat([legit_sample, fraud], axis=0)
```
